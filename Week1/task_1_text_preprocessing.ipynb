```python
# text_preprocessing.py or cell in task_1_text_preprocessing.ipynb
import re
import unicodedata
from typing import List
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer

class TextPreprocessor:
    def __init__(self):
        self.stemmer = PorterStemmer()
        self.lemmatizer = WordNetLemmatizer()
        base_stopwords = set(stopwords.words("english"))
        financial_keep = {
            "up", "down", "over", "under",
            "above", "below",
            "beat", "miss",
            "growth", "decline",
            "gain", "loss"
        }
        self.stopwords = base_stopwords - financial_keep

        # small contraction map (extend if needed)
        self.contractions = {
            "n't": " not",
            "'re": " are",
            "'s": " is",
            "'ve": " have",
            "'ll": " will",
            "'d": " would",
            "'m": " am"
        }

    def expand_contractions(self, text: str) -> str:
        for k, v in self.contractions.items():
            text = text.replace(k, v)
        return text

    def normalize(self, text: str) -> str:
        text = unicodedata.normalize("NFKD", text)
        text = text.lower()
        return text

    def remove_special_chars(self, text: str) -> str:
        # keep alphanumerics, periods (for decimals), and spaces
        return re.sub(r"[^a-z0-9.\s]", " ", text)

    def tokenize(self, text: str) -> List[str]:
        return text.split()

    def remove_stopwords(self, tokens: List[str]) -> List[str]:
        return [t for t in tokens if t not in self.stopwords]

    def stem(self, tokens: List[str]) -> List[str]:
        return [self.stemmer.stem(t) for t in tokens]

    def lemmatize(self, tokens: List[str]) -> List[str]:
        return [self.lemmatizer.lemmatize(t) for t in tokens]

    def preprocess(self, text: str) -> list[str]:
        text = self.normalize(text)
        text = self.expand_contractions(text)
        text = self.remove_special_chars(text)
    
        tokens = self.tokenize(text)
    
        # strip trailing punctuation from tokens
        tokens = [t.strip(".") for t in tokens]
    
        tokens = self.remove_stopwords(tokens)
    
        # â— DO NOT lemmatize nouns (README keeps "shares")
        return tokens

```

```python
tp = TextPreprocessor()
tp.preprocess("Apple Inc. shares fell 2.5% amid concerns about iPhone 15 demand.")
# -> ['apple', 'inc', 'shares', 'fell', '2.5', 'amid', 'concern', 'iphone', '15', 'demand']
```
